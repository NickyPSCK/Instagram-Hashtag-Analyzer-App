```python
X = df[['ImageID','path']]
y = df[['score']]
X_train, X_test, y_train, y_test = train_test_split(X, y, stratify = y, test_size=0.2, random_state=29)
```

```python
for index, row in X_test.iterrows():
  new_fol = str(y_test.loc[index, 'score'])
  new_path = '/content/flickr_new/validation/'+new_fol+"/"+str(row['ImageID'])+".jpg"
  os.rename(row['path'], new_path)
```

```python
img_height = 224
img_width = 224
batch_size = 64
```

```python
train_dir = '/content/flickr_new/train'
train_ds = tf.keras.preprocessing.image_dataset_from_directory(
  train_dir,
  seed=123,
  color_mode='rgb',
  image_size=(img_height, img_width),
  batch_size=batch_size)
```

```python
test_dir = '/content/flickr_new/validation'
test_ds = tf.keras.preprocessing.image_dataset_from_directory(
  test_dir,
  seed=123,
  color_mode='rgb',
  image_size=(img_height, img_width),
  batch_size=batch_size)
```

```python
class_names = train_ds.class_names
print(class_names)
```

## Model

```python
model = EfficientNetB0(weights = "imagenet", include_top=False, input_shape = (224, 224, 3))
model.summary()
```

```python
for i,layer in enumerate(model.layers):  
    print( f"Layer {i}: name={layer.name} , trainable={layer.trainable}=>{False}" )
    layer.trainable = False
```

```python
x = model.output

# Add our custom layer(s) to the end of the existing model 
x = GlobalAveragePooling2D()(x)
x = BatchNormalization()(x)
x = Dense(1024, activation="relu")(x)
x = Dropout(0.3)(x)
new_outputs = Dense(5, activation="softmax")(x)

# Create the final model 
model = Model(inputs=model.input, outputs=new_outputs)
model.summary()
```

### Train Model

```python
train_x = preprocess_input(train_ds)
test_x = preprocess_input(test_ds)
```

```python
# Set optimizer and loss
model.compile( loss="sparse_categorical_crossentropy", optimizer="adam", metrics=["acc"] ) 
```

```python
epochs=20
history = model.fit(
  train_ds,
  verbose=1,
  validation_data=test_ds,
  batch_size=128,
  epochs=epochs
)
```

```python
model.save('/content/drive/MyDrive/BADS7203 - Image/img_class')
model.save('/content/drive/MyDrive/BADS7203 - Image/img_class/img_class.h5')
```

```python
# Summarize history for accuracy
plt.plot(history.history['acc'])
plt.plot(history.history['val_acc'])
plt.title('Train accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()

# Summarize history for loss
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Train loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
```

```python
results = model.evaluate(test_ds, batch_size=128)
print("test loss, test acc:", results)
```

## Predict

### Function for prepare one image

```python
def crop_center_and_resize(img,size):
  width, height = img.shape[-2], img.shape[-3]
  if height > width :
    crop_width = width
    crop_height = width
  else:
    crop_width = height
    crop_height = height
  mid_x, mid_y = int(width/2), int(height/2)
  cw2, ch2 = int(crop_width/2), int(crop_height/2)
  crop_img = img[mid_y-ch2:mid_y+ch2, mid_x-cw2:mid_x+cw2]
  return cv2.resize(crop_img,(size, size))
```

```python
def create_image(image_path):
  image = cv2.imread(image_path)
  image = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
  image = crop_center_and_resize(image, 224)
  image = np.array(image)
  # image = image.reshape(1,image.shape[0],image.shape[1],image.shape[2])
  return image
```

```python
target_dict = ["Extreamly Negative","Negative","Neutal","Positive","Extreamly Positive"]
print(target_dict)
```
