```python
import pandas as pd 

def getInstagramUrlFromMediaId(media_id):
    media_id = int(media_id)
    alphabet = 'ABCDEFGHIJKLMNOPQRSTUVWXYZabcdefghijklmnopqrstuvwxyz0123456789-_'
    shortened_id = ''

    while media_id > 0:
        remainder = media_id % 64
        # dual conversion sign gets the right ID for new posts
        media_id = (media_id - remainder) // 64
        # remainder should be casted as an integer to avoid a type error. 
        shortened_id = alphabet[int(remainder)] + shortened_id

    return shortened_id

data1 = pd.read_csv('insta1_icassp2016_dataset.csv')
data2 = pd.read_csv('insta2_icassp2016_dataset.csv')
data3 = pd.read_csv('insta3_icassp2016_dataset.csv')
data = pd.concat([data1, data2, data3], axis=0,  ignore_index=True)

data['PureImageID'] = data['ImageID'].apply(lambda id:id.split('_')[0])


data['shortened_id'] = data['PureImageID'].apply(getInstagramUrlFromMediaId)
data['url'] = data['shortened_id'].apply(lambda shortened_id:'https://instagram.com/p/' + shortened_id + '/')

data.to_csv('icassp2016_dataset.csv', index=False)

print(data.info())


data.head()
```

```python
for short_code in tqdm(data['shortened_id']):
    try:
        post = Post.from_shortcode(il.context, short_code)
        il.filename_pattern=short_code
        il.download_post(post, target='icassp2016_dataset')
    except:
        with open('log.txt', 'a') as f:
            f.write(f'{short_code}\n')
```
